```python
from numpy.random import randn
import numpy as np
import os
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
import pandas as pd
from datetime import datetime
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('oms_pbosdbnums91_logsegment.txt', index_col=0, sep="|", parse_dates=True)
y_data = data['UsedLogsegment%']

y_data.plot(ax=ax, style='b-o')

crisis_data = [
#    (datetime(2007, 10, 11), 'Peak of bull market'),
#    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2017, 5, 4, 16, 52, 38), 'First Peak')
]

for date, label in crisis_data:
    #print(date)
    #print(int(y_data.asof(date)))
    #print(label)
    ax.annotate(label, xy=(date, int(y_data.asof(date)) + 2),
                xytext=(date, y_data.asof(date) + 8),
                arrowprops=dict(facecolor='red'),
                horizontalalignment='left', color='black', verticalalignment='top')

### Zoom in 
#ax.set_xlim(['2017-05-04 16:51:18', '2017-05-04 16:57:28'])
#ax.set_ylim([0, 18])

ax.set_title('NOMS Log Segment Usage')
ax.set_xlabel('Time')
ax.set_ylabel('% Used')
```
