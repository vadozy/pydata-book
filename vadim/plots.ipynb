```python
# Example 3.4 (Geometric distribution)
p = 0.3
def PX(n):
    return p * (1 - p)**(n - 1)

n = 20

plt.plot([PX(x) for x in range(1, n+1)], 'bo--')
```

```python
# Binomial distribution
n = 2
p = 0.5

def PX(k):
    return cb.nck(n,k) * (p**k) * (1 - p)**(n - k)

plt.plot([PX(x) for x in range(0, n+1)], 'bo--')
```
